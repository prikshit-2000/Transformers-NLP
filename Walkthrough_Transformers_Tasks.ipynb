```python
from transformers import pipeline
```

Sentiment Analysis

```python
classifier = pipeline('sentiment-analysis')
classifier([
            'That movie is too good',
             'That tastes dull'
             ])
```

Zero Shot Classification

```python
classifier = pipeline('zero-shot-classification')
classifier('Donald Trump lost the US elections in 2020',
           candidate_labels = ['politics','education','sports','news'])
```

Text Generation

```python
generator = pipeline('text-generation')
generator('Virat Kohli started his test career in 2008')
```

Distil gpt2

```python
generator= pipeline('text-generation',model='distilgpt2')
generator('Donald Trump lost the US elections in 2020',
          max_length=30,
    num_return_sequences=2)
```

unmasker

```python
unmasker= pipeline('fill-mask')
unmasker('Donald Trump lost the US <mask> in 2020',top_k = 2)
```

Named entity recognition

```python
ner = pipeline('ner',grouped_entities=True)
ner('Sundar pichai is the current CEO of Gooogle')
```

Q/A

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn"
)
```

Text Summarization
